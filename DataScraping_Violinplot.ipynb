#For equations of each event in groups

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
bf = []
gr = []
l = ['1991-92']
for i in l:
  for j in range(1,21):
    df=open(f'{i}_eq_{j}.txt','r')  # open file in READ_ONLY mode
    word = 'Fitness='
    for line in df:
      if word in line:
        a=line.split(word)
        bf.append(float(a[1]))
        gr.append(i)
        # print(float(a[1]))

df.close() # close file
```

```python
dict = {'group 7' : gr , 'best fitness' : bf }

df = pd.DataFrame(dict)
# df
plt.figure(figsize = (6,5))
ax = sns.violinplot(x= 'group 7', y='best fitness', data=df)
```
